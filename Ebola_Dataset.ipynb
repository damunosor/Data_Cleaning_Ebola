Data Cleaning for Ebola Dataset (Guinea,Liberia and Sierra Leona)

Dataset sources: https://apps.who.int/gho/data/node.ebola-sitrep.quick-downloads?lang=en

```python
import pandas as pd
import numpy as np
```

```python
Guinea=pd.read_csv('2016-05-11_Guinea.csv')
Liberia=pd.read_csv('2016-05-11_Liberia.csv')
Sierra_Leona=pd.read_csv('2016-05-11_Sierra_Leona.csv')
```

```python
Ebola_Countries = [Guinea, Liberia, Sierra_Leona]
Ebola= pd.concat(Ebola_Countries)
Ebola.shape
```

```python
Ebola.head(10)
```

```python
Ebola.tail(10)
```

```python
Ebola.info()

```

```python
# Undestanding the value count
print(Ebola.Country.value_counts(), '\n')
print(Ebola.Gender.value_counts(), '\n')
print(Ebola.Location.value_counts(), '\n')
print(Ebola['Ebola measure'].value_counts(), '\n')
print(Ebola['Case definition'].value_counts(), '\n')
print(Ebola['Ebola data source'].value_counts(), '\n')
print(Ebola['Epi week'].value_counts(), '\n')
print(Ebola['Indicator type'].value_counts(), '\n')


```

```python
Ebola.head()
```

```python
# Dropping columns with many missing values
Ebola.drop(columns=['Age group','Gender','Data as of','Outbreak identifier','Low','High','StdErr','StdDev','Comments'], inplace = True)
```

```python
Ebola.head(10)
```

```python
Ebola1 = pd.pivot_table(Ebola, values=['Numeric'], index=['Country', 'Epi week', 'Ebola data source'], columns=['Case definition'], aggfunc=np.sum)
Ebola1.head(300)
```

```python
Ebola2 = pd.DataFrame(Ebola1.to_records()) 
Ebola2.rename({'(\'Numeric\', \'Confirmed\')' :  'Confirmed_cases', '(\'Numeric\', \'Probable\')' :  'Probable_cases', }, axis = 1, inplace = True)
Ebola2.head(300)
```

```python
Ebola2.shape

```

```python
Ebola2.to_csv('Ebola_Countries_pivoted.csv')
```
